```python
def assigned_centroid(iris_df,centroids):
    assigned_centroid = []
    for i in  iris_df:
        distance=[]
        for j in centroids:
             distance.append(calc_distance(i, j))
        assigned_centroid.append(np.argmin(distance))#Returns the indices of the minimum values along an axis.
    return assigned_centroid
    
    
    
```

```python
def calc_new_centroids(iris_df,assigned_centroid):
    new_centroids = []
    new_df = pd.concat([pd.DataFrame(iris_df), pd.DataFrame(assigned_centroid, columns=['cluster'])],axis=1)
    for c in set(new_df['cluster']):
        current_cluster = new_df[new_df['cluster'] == c][new_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        new_centroids.append(cluster_mean)
    return new_centroids
    
```

```python
from scipy.spatial.distance import cdist 
def k_means(iris_df,k):
    init_centroids = random.sample(range(0, len(iris_df)), k)
    #Randomly choosing Centroids 
    centroids = iris_df[init_centroids, :] 
    distances = cdist(iris_df, centroids ,'euclidean')
    points = np.array([np.argmin(i) for i in distances]) #retaurn index of mini destance
    #Repeating the above steps for a defined number of iterations
    
    for o in range(100): 
        centroids = []
        for idx in range(k):
            #Updating Centroids by taking mean of Cluster it belongs to
            temp_cent = iris_df[points==idx].mean(axis=0) 
            centroids.append(temp_cent)
 
        centroids = np.vstack(centroids) #Updated Centroids 
         
        distances = cdist(iris_df, centroids ,'euclidean')
        points = np.array([np.argmin(j) for j in distances])
        
         
    return points , centroids
    
    
    
```

```python
label,center_point =  k_means(iris_df,4)
label
```

# # show The elbow method

```python


K = range(1, 10)
dist_List = [] 
for k in K:
    # Building and fitting the model
    label,center_point =  k_means(iris_df,k)
    dist_List.append(sum(np.min(cdist(iris_df,center_point,'euclidean'), axis=1)) / iris_df.shape[0])
plt.plot(K, dist_List)
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
    
   
```
